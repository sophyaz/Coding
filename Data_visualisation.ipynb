### 1. Boxplot

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset into a DataFrame
df = pd.read_csv('Cars93.csv')

# Filter data for specific car manufacturers: Audi, Hyundai, Suzuki, and Toyota
selected_manufacturers = ['Audi', 'Hyundai', 'Suzuki', 'Toyota']
filtered_data = df[df['Manufacturer'].isin(selected_manufacturers)]

# Create a box plot for revs per mile for the selected manufacturers
plt.figure(figsize=(10, 6))  # Set the figure size
plt.boxplot([filtered_data[filtered_data['Manufacturer'] == manufacturer]['Rev.per.mile'] for manufacturer in selected_manufacturers],
            labels=selected_manufacturers)

# Add labels and title
plt.xlabel('Manufacturer')
plt.ylabel('Rev.per.mile')
plt.title('Rev.per.mile Distribution for Selected Manufacturers')

# Show plot
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.grid(True)
plt.show()

# Find the manufacturer with the highest revs per mile
highest_revs_manufacturer = filtered_data.loc[filtered_data['Rev.per.mile'].idxmax(), 'Manufacturer']
print("Manufacturer with the highest Rev.per.mile:", highest_revs_manufacturer)
```

**Boxplot question:**  
Which of these manufacturers has the car with the highest revs per mile?


**Boxplot answer:**
From the boxplot we can see that Suzuki has the highest revs per mile 

```python
import matplotlib.pyplot as plt
import pandas as pd
car_data = pd.read_csv('Cars93.csv')
# Create a figure and axis object
fig, ax = plt.subplots()

# Plot histograms for City MPG and Highway MPG
ax.hist(car_data['MPG.city'], bins=20, alpha=0.5, label='City MPG')
ax.hist(car_data['MPG.highway'], bins=20, alpha=0.5, label='Highway MPG')

# Add labels and legend
ax.set_xlabel('MPG')
ax.set_ylabel('Frequency')
ax.set_title('Distribution of MPG in City and Highway')
ax.legend()

# Show plot
plt.show()

```

**Histogram question:**  
Is it generally more fuel efficient to drive in the city or on the highway?

**Histogram answer:**
From the graph we can see it is more efficient to drive in the city as there are less MPG than on the highway 

### 3. Lineplot

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset into a DataFrame
# Replace 'cars.csv' with the actual filename and path if it's different
df = pd.read_csv('Cars93.csv')

# Create a line plot for 'wheelbase' vs 'turning circle'
plt.plot(df['Wheelbase'], df['Turn.circle'], marker='o', linestyle='-')

# Add labels and title
plt.xlabel('Wheelbase')
plt.ylabel('Turn.circle')
plt.title('Relationship between Wheelbase and Turn.circle')

# Show plot
plt.grid(True)
plt.show()
```

### 4. Barplot

```python
# Barplot for mean horsepower
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset into a DataFrame
# Replace 'cars.csv' with the actual filename and path if it's different
df = pd.read_csv('Cars93.csv')

# Calculate mean horsepower for each car type
mean_hp_by_type = df.groupby('Type')['Horsepower'].mean().sort_values()

# Create a bar plot for mean horsepower by car type
mean_hp_by_type.plot(kind='bar', color='skyblue')

# Add labels and title
plt.xlabel('Car Type')
plt.ylabel('Mean Horsepower')
plt.title('Mean Horsepower by Car Type')

# Show plot
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.tight_layout()  # Adjust layout to prevent label cutoff
plt.show()
```

**Barplot question:**  
Does a larger car mean more horsepower?



**Barplot answer:**
From the barplot we can see that having a larger car type results in higher mean horsepower.
